```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

# Make data
u = np.linspace(0, 2 * np.pi, 100)  # 要素数を増やす
v = np.linspace(0, np.pi, 100)  # 要素数を増やす
U, V = np.meshgrid(u, v)  # uとvから行列を作成
x = 10 * np.cos(U) * np.sin(V)  # UとVを直接使用
y = 10 * np.sin(U) * np.sin(V)  # UとVを直接使用
z = 10 * np.cos(V)  # Uは必要ない

# u = np.linspace(0, 2 * np.pi, 100)
# v = np.linspace(0, np.pi, 100)
# x = 10 * np.outer(np.cos(u), np.sin(v))
# y = 10 * np.outer(np.sin(u), np.sin(v))
# z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))

colors = plt.cm.jet((x - x.min()) / (x.max() - x.min()))  # uの値に基づく色


# Plot the surface
# ax.plot_surface(x, y, z, shade=False)  # type: ignore
ax.plot_surface(x, y, z, shade=True, rstride=1, cstride=1, facecolors=colors)


# Set an equal aspect ratio
ax.set_aspect("equal")
```
